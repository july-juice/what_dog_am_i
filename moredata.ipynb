```python
import cv2
import os
import time

save_path = input("Enter folder save path: ").strip()
prefix = input("Enter the file prefix: ").strip()

# Make sure folder exists

os.makedirs(save_path, exist_ok = True)

interval = 2

# Load camera

cam = cv2.VideoCapture(1)

if not cam.isOpened():
    print("Camera nuh work breddah")
    exit()

print(f"\n Camera work. Taking photos every {interval} seconds. Press 'q' to quit.")

counter = 1
misfire_count = 0
start_time = time.time()

while True:
    ret,frame = cam.read()
    if not ret:
        print("Frame didn't capture")
        misfire_count +=1
        continue
    if misfire_count == 5:
        print("Camera nuh work breddah. Mi try hard.")
        exit()

    cv2.imshow("Dataset additions", frame)

    if time.time() - start_time >= interval:
        filename = f"{prefix}{counter:03d}.jpg"
        filepath = os.path.join(save_path, filename)
        cv2.imwrite(filepath, frame)
        print(f"Saved: {filepath}")
        counter += 1
        start_time = time.time()

    # Quit if q pressed

    if counter == 21:
        print("NV")
        break

    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("NV")
        break

cam.release()
cv2.destroyAllWindows()
```
